```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

fig, ((ax, ax1,ax2),(ax3,ax4,ax5)) = plt.subplots(2,3, figsize=(16,10))
fig.suptitle('sistemas de color', fontsize=20)
fig1, (i1,i2,i3) = plt.subplots(1,3, figsize=(16,6))
fig1.suptitle('canales RGB', fontsize=20)
fig2, (HSV1,HSV2,HSV3) = plt.subplots(1,3, figsize=(16,6))
fig2.suptitle('canales HSV', fontsize=20)
fig3, (LAB1,LAB2,LAB3) = plt.subplots(1,3, figsize=(16,6))
fig3.suptitle('canales LAB', fontsize=20)
fig4, (YUV1,YUV2,YUV3) = plt.subplots(1,3, figsize=(16,6))
fig4.suptitle('canales YUV', fontsize=20)
fig5, (XYZ1,XYZ2,XYZ3) = plt.subplots(1,3, figsize=(16,6))
fig5.suptitle('canales XYZ', fontsize=20)
fig6, (A1,A2,A3) = plt.subplots(1,3, figsize=(16,6))
fig6.suptitle('escala de grises', fontsize=20)


img = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ax.imshow(img,aspect='auto')
ax.set_title('SISTEMA RGB')
imgr=img[:,:,0]
imgg=img[:,:,1]
imgb=img[:,:,2]
i1.imshow(imgr,aspect='auto',cmap='Reds')
i1.set_title('sistema RGB canal R')
i2.imshow(imgg,aspect='auto',cmap='Greens')
i2.set_title('sistema RGB canal G')
i3.imshow(imgb,aspect='auto',cmap='Blues')
i3.set_title('sistema RGB canal B')

img1 = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
ax1.imshow(img1,aspect='auto')
ax1.set_title('SISTEMA HSV')
imgH=img1[:,:,0]
imgS=img1[:,:,1]
imgV=img1[:,:,2]
HSV1.imshow(imgH,aspect='auto',cmap='gray')
HSV1.set_title('sistema HSV canal H')
HSV2.imshow(imgS,aspect='auto',cmap='gray')
HSV2.set_title('sistema HSV canal S')
HSV3.imshow(imgV,aspect='auto',cmap='gray')
HSV3.set_title('sistema HSV canal V')

img2 = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2LAB)
ax2.imshow(img2,aspect='auto')
ax2.set_title('SISTEMA LAB')
imgL=img2[:,:,0]
imgA=img2[:,:,1]
imgB=img2[:,:,2]
LAB1.imshow(imgL,aspect='auto',cmap='gray')
LAB1.set_title('sistema LAB canal L')
LAB2.imshow(imgA,aspect='auto',cmap='gray')
LAB2.set_title('sistema LAB canal A')
LAB3.imshow(imgB,aspect='auto',cmap='gray')
LAB3.set_title('sistema LAB canal B')

img3 = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2YUV)
ax3.imshow(img3,aspect='auto')
ax3.set_title('SISTEMA YUV')
imgY=img3[:,:,0]
imgU=img3[:,:,1]
imgV=img3[:,:,2]
YUV1.imshow(imgY,aspect='auto',cmap='gray')
YUV1.set_title('sistema YUV canal Y')
YUV2.imshow(imgU,aspect='auto',cmap='gray')
YUV2.set_title('sistema YUV canal U')
YUV3.imshow(imgV,aspect='auto',cmap='gray')
YUV3.set_title('sistema YUV canal V')

img4 = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2XYZ)
ax4.imshow(img4,aspect='auto')
ax4.set_title('SISTEMA XYZ')
imgX1=img4[:,:,0]
imgX2=img4[:,:,1]
imgX3=img4[:,:,2]
XYZ1.imshow(imgX1,aspect='auto',cmap='gray')
XYZ1.set_title('sistema XYZ canal X')
XYZ2.imshow(imgX2,aspect='auto',cmap='gray')
XYZ2.set_title('sistema XYZ canal Y')
XYZ3.imshow(imgX3,aspect='auto',cmap='gray')
XYZ3.set_title('sistema XYZ canal Z')

img5 = cv2.imread("res/frutas.jpg", cv2.IMREAD_COLOR)
img5 = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)
img6=  cv2.imread("res/frutas.jpg", cv2.IMREAD_GRAYSCALE)
ax5.imshow(img6,aspect='auto',cmap='gray')
ax5.set_title('ESCALA DE GRISES')
imgY1=img5[:,:,0]
imgY2=img5[:,:,1]
imgY3=img5[:,:,2]
A1.imshow(imgY1,aspect='auto',cmap='gray')
A1.set_title('escala de grises canal R')
A2.imshow(imgY2,aspect='auto',cmap='gray')
A2.set_title('escala de grises canal G')
A3.imshow(imgY3,aspect='auto',cmap='gray')
A3.set_title('escala de grises canal B')
```
